```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data
import seaborn as sns
```

```python
# Use the iris database to answer the following questions.
iris = sns.load_dataset('iris')
sns.pairplot(iris)
```

```python
# What does the distribution of petal lengths look like?
sns.histplot(iris.petal_length)
```

```python
# Is there a correlation between petal length and petal width?
sns.relplot(x='petal_length', y='petal_width', data=iris)
```

```python
# Would it be reasonable to predict species based on sepal width and sepal length?
sns.relplot(x='sepal_length', y='sepal_width', col='species', data=iris)
```

```python
# Which features would be best used to predict species?
sns.relplot(x='petal_width', y='sepal_width', hue='species', data=iris)
sns.relplot(x='petal_length', y='sepal_length', hue='species', data=iris)
```

```python
# Using the lesson as an example, use seaborn's load_dataset function to load the 
# anscombe data set. 
ans = sns.load_dataset('anscombe')
ans.sample(10)
```

```python
# Use pandas to group the data by the dataset column, and 
# calculate summary statistics for each dataset. What do you notice?
x_agg = ans.groupby('dataset').x.agg(['mean', 'median', 'count'])
y_agg = ans.groupby('dataset').y.agg(['mean', 'median', 'count'])
xy_agg = x_agg.merge(y_agg, on='dataset')
xy_agg
```

```python
# Plot the x and y values from the anscombe data. 
# Each dataset should be in a separate column.
sns.relplot(x='x', y='y', col='dataset', data=ans)
```

```python
# Load the swiss dataset and read it's documentation. 
data('swiss', show_doc=True)
swiss = data('swiss')
swiss
```

```python
# Create visualizations to answer the following questions:
# Create an attribute named is_catholic that holds a boolean value of 
# whether or not the province is Catholic. 
# (Choose a cutoff point for what constitutes catholic)
is_catholic = swiss['Catholic'] >= 50
is_catholic.head()
```

```python
# Does whether or not a province is Catholic influence fertility?
swiss['is_catholic'] = is_catholic
sns.relplot(x='Catholic', y='Fertility', hue='is_catholic', data=swiss)
```

```python
# What measure correlates most strongly with fertility?
# swiss = swiss.drop(columns='is_catholic')
sns.regplot(x='Agriculture', y='Fertility', data=swiss, label='Agriculture')
sns.regplot(x='Examination', y='Fertility', data=swiss, label='Examination')
sns.regplot(x='Education', y='Fertility', data=swiss, label='Education')
sns.regplot(x='Catholic', y='Fertility', data=swiss, label='Catholic')
plt.legend(loc='lower right')
```

```python
# Using the chipotle dataset from the previous exercise
from env import host, username, password

def get_db_url(username, hostname, password, db_name):
    return f'mysql+pymysql://{username}:{password}@{hostname}/{db_name}'

url = get_db_url(username, host, password, 'chipotle')
chip = pd.read_sql('SELECT * FROM orders', url)
chip.sample(10)
```

```python
# Create a bar chart that shows the 4 most popular items and the revenue produced by each.
chip['item_price'] = chip['item_price'].str.lstrip('$')

chip = chip.astype({'item_price':'float64'})

top_4 = chip.groupby("item_name").sum().sort_values(by=["quantity"], ascending = False).head(4)
top_4 = top_4.drop(columns=['id', 'order_id', 'quantity'])

top_4.plot.barh()
```
