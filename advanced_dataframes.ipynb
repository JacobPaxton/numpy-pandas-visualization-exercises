```python
# Load the mpg dataset from PyDataset.
# Output and read the documentation for the mpg dataset.
from pydataset import data
mpg = data('mpg')
data('mpg', show_doc=True)
```

```python
# How many rows and columns are in the dataset?
mpg.shape
```

```python
# Check out your column names and perform any cleanup you may want on them.
mpg.columns
mpg = mpg.rename(columns={'cty': 'city'})
mpg
```

```python
# Display the summary statistics for the dataset.
mpg.info()
mpg.describe()
```

```python
# How many different manufacturers are there?
len(np.unique(mpg.manufacturer))
```

```python
# How many different models are there?
len(np.unique(mpg.model))
```

```python
# Create a column named mileage_difference like you did in the DataFrames exercises;
# this column should contain the difference between highway and city mileage for each car.
mpg['mileage_difference'] = mpg.hwy - mpg.city
mpg
```

```python
# Create a column named average_mileage like you did in the DataFrames exercises;
# this is the mean of the city and highway mileage.
mpg['average_mileage'] = (mpg.city + mpg.hwy) / 2
mpg
```

```python
# Create a new column on the mpg dataset named is_automatic that 
# holds boolean values denoting whether the car has an automatic transmission.
mpg['is_automatic'] = mpg.trans.str.startswith('auto')
mpg
```

```python
# Using the mpg dataset, find out which which manufacturer has 
# the best miles per gallon on average?
mpg.groupby('manufacturer').average_mileage.mean().sort_values(ascending=False)
```

```python
# Do automatic or manual cars have better miles per gallon?
auto_mpg = round(mpg[mpg.is_automatic].average_mileage.mean(),2)
manual_mpg = round(mpg[~mpg.is_automatic].average_mileage.mean(),2)
print("Auto is better than manual on gas consumption:", (auto_mpg > manual_mpg))
```

```python
# Use your get_db_url function to help you explore the data from the chipotle database.
url = get_db_url(username, host, password, 'chipotle')
chip = pd.read_sql('SELECT * FROM orders', url)
chip.sample(10)
```

```python
# What is the total price for each order?
chip['price_numeric'] = chip['item_price'].str.lstrip('$')
chip = chip.astype({'price_numeric':'float64'})
each_order_price = chip.groupby('order_id').price_numeric.sum()
each_order_price
```

```python
# What are the most popular 3 items?
chip.groupby('item_name').quantity.sum().sort_values(ascending=False).head(3)
```

```python
# Which item has produced the most revenue?
chip.groupby('item_name').price_numeric.sum().sort_values(ascending=False).head(2)
```

```python
# Join the employees and titles DataFrames together.
emps_ttls = emps.merge(ttls, on='emp_no', how='inner')
emps_ttls
```

```python
# For each title, 
# find the hire date of the employee that was hired most recently with that title.
emps_ttls.groupby('title').hire_date.max()
```

```python
# Write the code necessary to 
# create a cross tabulation of the number of titles by department. 
# (Hint: this will involve a combination of SQL code to pull the necessary data and 
# python/pandas code to perform the manipulations.)

# Pull dept_emp table
url = get_db_url(username, host, password, 'employees')
dept_emp = pd.read_sql('SELECT * FROM dept_emp', url)
dept_emp.sample(10)
```
