```python
import numpy as np
import pandas as pd
from scipy.special import factorial
import matplotlib.pyplot as plt
```

```python
# Title your chart "Big O Notation"
plt.title("Big O Notation")
# Label your x axis "Elements"
plt.xlabel("Elements")
# Label your y axis "Operations"
plt.ylabel("Operations")
# Label your curves or make a legend for the curves
# Use LaTex notation where possible

n = pd.Series(np.arange(1, 3.2, 0.2))
# Curves to graph

# y = 0n +1
# and label the curve "O(1)"
plt.plot(n, (0 * n) + 1, label=r'O(1)')

# y = log(n)
# and label the curve "O(log n)"
plt.plot(n, np.log(n), label=r'O(log(n))')

# y = n
# and label the curve "O(n)"
plt.plot(n, n, label=r'O(n)')

# y = n * log(n)
#  and label it "O(n log n)"
plt.plot(n, (n * np.log(n)), label=r'O(n * log(n))')

# y = n^2
# and label it "O(n^2)"
plt.plot(n, n**2, label=r'O(n^2)')

# y = 2^n
# and label it "O(2^n)"
plt.plot(n, 2**n, label=r'O(2^n)')

# y = n!
# and label it "O(n!)"
plt.plot(n, factorial(n), label=r'O(n!)')

# y = n^n
#  and label it "O(n^n)"
plt.plot(n, n**n, label=r'O(n^n)')

plt.legend()
plt.show()
```
