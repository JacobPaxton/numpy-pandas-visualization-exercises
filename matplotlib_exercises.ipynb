```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Use matplotlib to plot the following equation:
# y = x^2 − x + 2
# You'll need to write the code that generates the x and y points.
def x_squared_minus_x_plus_2(num):
    return (num**2) - num + 2

series_of_nums = pd.Series(range(-10, 10))
y = series_of_nums.apply(x_squared_minus_x_plus_2)
plt.plot(y)
plt.xlim(-5, 20)

# Add an anotation for the point 0, 0, the origin.
plt.annotate('(0, 0)', xy=(0, 0), xytext=(1, 20), arrowprops={'color':'black'})
plt.show()
```

```python
# Create and label 4 separate charts for the following equations 
# (choose a range for x that makes sense):
# You can use functions from the math module to help implement some of the equations above.
# y = √x
def sqrt_x(num):
    return num ** 0.5

series_of_nums = pd.Series(range(0, 101))
y = series_of_nums.apply(sqrt_x)
plt.plot(series_of_nums, y, c='red')
plt.title(r'$\sqrt{x}$')
```

```python
# y = x^3
def x_cubed(num):
    return num ** 3

series_of_nums = pd.Series(range(-5, 11))
y = series_of_nums.apply(x_cubed)
plt.plot(series_of_nums, y, c='green')
plt.title(r'$x^3$')
```

```python
# y = 2^x
def two_to_power_of_x(num):
    return 2 ** num

series_of_nums = pd.Series(range(-5, 6))
y = series_of_nums.apply(two_to_power_of_x)
plt.plot(series_of_nums, y, c='blue')
plt.title(r'$2^x$')
```

```python
# y = 1/(x+1)
series_of_nums = pd.Series(range(-10, 11))
y = 1 / (series_of_nums + 1)
plt.plot(series_of_nums, y, c='orange')
plt.title(r'$\frac{1}{x+1}$')
```

```python
# Combine the figures you created in the last step into one large figure with 4 subplots.
plt.subplot(221)
series_of_nums = pd.Series(range(0, 101))
y = series_of_nums.apply(sqrt_x)
plt.plot(series_of_nums, y, c='red')
plt.title(r'$\sqrt{x}$')

plt.subplot(222)
series_of_nums = pd.Series(range(-5, 11))
y = series_of_nums.apply(x_cubed)
plt.plot(series_of_nums, y, c='green')
plt.title(r'$x^3$')

plt.subplot(223)
series_of_nums = pd.Series(range(-5, 6))
y = series_of_nums.apply(two_to_power_of_x)
plt.plot(series_of_nums, y, c='blue')
plt.title(r'$2^x$')

plt.subplot(224)
series_of_nums = pd.Series(range(-10, 11))
y = 1 / (series_of_nums + 1)
plt.plot(series_of_nums, y, c='orange')
plt.title(r'$\frac{1}{x+1}$')

plt.suptitle('Math Equations, Plotted (Subplots)')
plt.tight_layout()
plt.show()
```

```python
# Combine the figures you created in the last step into one figure where 
# each of the 4 equations has a different color for the points. 
# Be sure to include a legend and an appropriate title for the figure.
series_of_nums = pd.Series(np.arange(0, 3, 0.25))
y = series_of_nums.apply(sqrt_x)
plt.plot(series_of_nums, y, c='red', label=r'$\sqrt{x}$')

series_of_nums = pd.Series(np.arange(0, 3, 0.25))
y = series_of_nums.apply(x_cubed)
plt.plot(series_of_nums, y, c='green', label=r'$x^3$')

series_of_nums = pd.Series(np.arange(0, 3, 0.25))
y = series_of_nums.apply(two_to_power_of_x)
plt.plot(series_of_nums, y, c='blue', label=r'$2^x$')

series_of_nums = pd.Series(np.arange(0, 3, 0.25))
y = 1 / (series_of_nums + 1)
plt.plot(series_of_nums, y, c='orange', label=r'$\frac{1}{x+1}$')

plt.title('Math Equations, Plotted (Combined)')
plt.legend()
plt.show()
```
